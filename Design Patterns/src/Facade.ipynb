# Facade

Facade Method Design Pattern provides a unified interface to a set of interfaces in a subsystem.

Facade defines a high-level interface that makes the
subsystem easier to use.

## Components of the Home Theater System

First, let's define the individual components of the home theater system:

```python
class DVDPlayer:
    def on(self):
        print("DVD Player is on")

    def off(self):
        print("DVD Player is off")

    def play(self, movie):
        print(f"Playing '{movie}'")

class Amplifier:
    def on(self):
        print("Amplifier is on")

    def off(self):
        print("Amplifier is off")

    def set_volume(self, level):
        print(f"Amplifier volume set to {level}")

class Projector:
    def on(self):
        print("Projector is on")

    def off(self):
        print("Projector is off")

    def wide_screen_mode(self):
        print("Projector in wide screen mode")
```

## The Facade

Next, we'll create the Facade class that provides a simplified interface to control these components:

```python
class HomeTheaterFacade:
    def __init__(self, dvd_player, amplifier, projector):
        self.dvd_player = dvd_player
        self.amplifier = amplifier
        self.projector = projector

    def watch_movie(self, movie):
        print("Get ready to watch a movie...")
        self.dvd_player.on()
        self.amplifier.on()
        self.projector.on()
        self.projector.wide_screen_mode()
        self.amplifier.set_volume(10)
        self.dvd_player.play(movie)

    def end_movie(self):
        print("Shutting movie theater down...")
        self.dvd_player.off()
        self.amplifier.off()
        self.projector.off()
```

## Using the Facade

Finally, let's see how we can use the Facade to control the home theater system:

```python
if __name__ == "__main__":
    # Creating instances of components
    dvd_player = DVDPlayer()
    amplifier = Amplifier()
    projector = Projector()

    # Creating the Facade
    home_theater = HomeTheaterFacade(dvd_player, amplifier, projector)

    # Using the Facade to watch a movie
    home_theater.watch_movie("Inception")

    # Using the Facade to end the movie
    home_theater.end_movie()
```
